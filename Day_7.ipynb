## 데이터 불러오기 및 전처리

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings(action='ignore')
```

```python
population = pd.read_excel('07_population_raw_data.xlsx', header=1)
# NaN 값을 다른 값으로 채우기(pad = 앞에 있는 데이터 값을 가져와서 채워라)
population.fillna(method='pad', inplace=True)
population
```

```python
population.info()
```

```python
# 컬럼 이름 변경
population.rename(
    columns={
        "행정구역(동읍면)별(1)": "광역시도",
        "행정구역(동읍면)별(2)": "시도",
        "계": "인구수"
    }, inplace=True
)
population.head()
```

```python
# 소계 제거 
population = population[population["시도"] != "소계"]
population.head()
```

```python
population.rename(
    columns={"항목": "구분"}, inplace=True
)
population.head()
```

```python
# 구분 컬럼에서 총인구수 (명), 남자인구수 (명), 여자인구수 (명)을 합계, 남자, 여자로 변경
population.loc[population["구분"] == "총인구수 (명)", "구분"] = "합계"
population.loc[population["구분"] == "남자인구수 (명)", "구분"] = "남자"
population.loc[population["구분"] == "여자인구수 (명)", "구분"] = "여자"

population
```

```python
# 소멸지역 조사를 위한 데이터 정리
population["20-39세"] = (
    population["20 - 24세"] + population["25 - 29세"] + population["30 - 34세"] + population["35 - 39세"]
)

population["65세이상"] = (
    population["65 - 69세"] + population["70 - 74세"] + population["75 - 79세"] + population["80 - 84세"] + population["85 - 89세"] + population["90 - 94세"] + population["95 - 99세"] + population["100+"]
)
```

```python
population
```

```python
# pivot_table 
pop_df = pd.pivot_table(
    data=population,
    index=["광역시도", "시도"], 
    columns=["구분"],
    values=["인구수", "20-39세", "65세이상"]
)
pop_df
```

```python
# 소멸 비율 계산 
pop_df["소멸비율"] = pop_df["20-39세", "여자"] / (pop_df["65세이상", "합계"] / 2)
pop_df.head()
```

```python
# 소멸위기지역 컬럼 생성 
pop_df["소멸위기지역"] = pop_df["소멸비율"] < 1.0 
pop_df
```

```python
# 소멸위기지역 조회 
pop_df[pop_df["소멸위기지역"] == True].index.get_level_values(1)
```

```python
pop_df.reset_index(inplace=True)
pop_df.head()
```

```python
# 컬럼 이름을 재지정해서 정리
tmp_columns = [
    pop_df.columns.get_level_values(0)[n] + pop_df.columns.get_level_values(1)[n]
    for n in range(0, len(pop_df.columns.get_level_values(0)))
]

pop_df.columns = tmp_columns
pop_df.head()
```

```python
tmp_gu_dict = {
    "수원": ["장안구", "권선구", "팔달구", "영통구"],
    "성남": ["수정구", "중원구", "분당구"],
    "안양": ["만안구", "동안구"],
    "안산": ["상록구", "단원구"],
    "고양": ["덕양구", "일산동구", "일산서구"],
    "용인": ["처인구", "기흥구", "수지구"],
    "청주": ["상당구", "서원구", "흥덕구", "청원구"],
    "천안": ["동남구", "서북구"],
    "전주": ["완산구", "덕진구"],
    "포항": ["남구", "북구"],
    "창원": ["의창구", "성산구", "진해구", "마산합포구", "마산회원구"],
    "부천": ["오정구", "원미구", "소사구"],
}
```

#### (1) 일반 시 이름과 세종시, 광역시도 일반 구 정리

```python
# 서울특별시 -> 서울, 세종특별자치시 -> 세종, 중구 -> 서울 중구 이런식으로 정리해서 si_name 리스트에 담기
for idx, row in pop_df.iterrows():
    if row['광역시도'][-3:] not in ['광역시', '특별시', '자치시']:
        si_name[idx] = row['시도'][:-1]
    elif row['광역시도'] == '세종특별자치시':
        si_name[idx] = '세종'
    else:
        if len(row["시도"]) == 2:
            si_name[idx] = row["광역시도"][:2] + " " + row["시도"]
        else:
            si_name[idx] = row["광역시도"][:2] + " " + row["시도"][:-1]
```

#### (2) 행정구

```python
# 앞에서 만들었던 딕셔너리의 값을 이용해서 장안구 -> 수원 장안 이런 식으로 정리
for idx, row in pop_df.iterrows():
    if row["광역시도"][-3:] not in ["광역시", "특별시", "자치시"]:
        for keys, values in tmp_gu_dict.items():
            if row["시도"] in values:
                if len(row["시도"]) == 2:
                    si_name[idx] = keys + " " + row["시도"]
                    
                elif row["시도"] in ["마산합포구", "마산회원구"]:
                    si_name[idx] = keys + " " + row["시도"][2:-1]
                
                else:
                    si_name[idx] = keys + " " + row["시도"][:-1]
```

```python
si_name
```

#### (3) 고성군

```python
# 강원도와 경상남도에 고성이라는 같은 이름의 지역이 있는 것을 구분
for idx, row in pop_df.iterrows():
    if row["광역시도"][-3:] not in ["광역시", "특별시", "자치시"]:
        if row["시도"][:-1] == "고성" and row["광역시도"] == "강원도": 
            si_name[idx] = "고성(강원)"
        elif row["시도"][:-1] == "고성" and row["광역시도"] == "경상남도": 
            si_name[idx] = "고성(경남)"
```

```python
# 시각화한 그림에 text를 달아주는 함수
def plot_text_simple(draw_korea):
    for idx, row in draw_korea.iterrows():
        # 서울 중구 -> 이렇게 작성되어 있는 것은 줄을 띄우고 작성
        if len(row["ID"].split()) == 2:
            dispname = "{}\n{}".format(row["ID"].split()[0], row["ID"].split()[1])
        # 고성이 다른 지역에 2개가 있지만 지도에서는 구분이 가능하므로 그냥 고성이라 표현
        elif row["ID"][:2] == "고성":
            dispname = "고성"
        # 그냥 그대로 표현
        else:
            dispname = row["ID"]

        # 한 라인에 세글자 이상인 것은 글씨 크기를 줄이는 과정    
        if len(dispname.splitlines()[-1]) >= 3:
            fontsize, linespacing = 9.5, 1.5
        else:
            fontsize, linespacing = 11, 1.2

        # matplotlib의 주석을 다는 함수
        plt.annotate(
            dispname,
            (row["x"] + 0.5, row["y"] + 0.5),
            weight="bold",
            fontsize=fontsize,
            linespacing=linespacing,
            ha="center", # 수평 정렬
            va="center", # 수직 정렬 
        )
```

```python
# 본격적으로 시각화를 하는 함수
def simpleDraw(draw_korea):
    plt.figure(figsize=(8, 11))
    
    plot_text_simple(draw_korea)
    
    for path in lines:
        # 각각의 경계선의 x좌표, y좌표를 따로 모아둠
        ys, xs = zip(*path)
        plt.plot(xs, ys, c="black", lw=1.5)
    
    # 현재 그래픽의 y축 뒤집기
    plt.gca().invert_yaxis()
    # 플롯 영역의 축을 숨기기
    plt.axis("off")
    # 그래픽 요소들을 자동으로 조정하여 플롯을 더 보기 좋게 만들기
    plt.tight_layout()
    plt.show()
```

#### 그림을 그리기 위한 데이터를 계산하는 함수 
- 색상을 만들 때, 최소값을 흰색 
- blockedMap: 인구현황(pop_df) 
- targetData: 그리고 싶은 컬럼

```python
def get_data_info(targetData, blockedMap):
    whitelabelmin = (
        max(blockedMap[targetData]) - min(blockedMap[targetData])
    ) * 0.25 + min(blockedMap[targetData])
    vmin = min(blockedMap[targetData])
    vmax = max(blockedMap[targetData])
    
    mapdata = blockedMap.pivot_table(index="y", columns="x", values=targetData)

    return mapdata, vmax, vmin, whitelabelmin
```

```python
# 여기서는 중간값을 흰색
def get_data_info_for_zero_center(targetData, blockedMap):
    whitelabelmin = 5 
    tmp_max = max(
        [np.abs(min(blockedMap[targetData])), np.abs(max(blockedMap[targetData]))]
    )
    vmin, vmax = -tmp_max, tmp_max
    mapdata = blockedMap.pivot_table(index="y", columns="x", values=targetData)
    return mapdata, vmax, vmin, whitelabelmin
```

```python
# 위에서 만들었던 함수에서 약간 변경
def plot_text(targetData, blockedMap, whitelabelmin):
    for idx, row in blockedMap.iterrows():
        if len(row["ID"].split()) == 2:
            dispname = "{}\n{}".format(row["ID"].split()[0], row["ID"].split()[1])
        elif row["ID"][:2] == "고성":
            dispname = "고성"
        else:
            dispname = row["ID"]
            
        if len(dispname.splitlines()[-1]) >= 3:
            fontsize, linespacing = 9.5, 1.5
        else:
            fontsize, linespacing = 11, 1.2


        annocolor = "white" if np.abs(row[targetData]) > whitelabelmin else "black"
        plt.annotate(
            dispname,
            (row["x"] + 0.5, row["y"] + 0.5),
            weight="bold",
            color=annocolor,
            fontsize=fontsize,
            linespacing=linespacing,
            ha="center", # 수평 정렬
            va="center", # 수직 정렬 
        )
```

```python
def drawKorea(targetData, blockedMap, cmapname, zeroCenter=False):
    if zeroCenter:
        masked_mapdata, vmax, vmin, whitelabelmin = get_data_info_for_zero_center(targetData, blockedMap)
    
    if not zeroCenter:
        masked_mapdata, vmax, vmin, whitelabelmin = get_data_info(targetData, blockedMap)
        
    plt.figure(figsize=(8, 11))
    plt.pcolor(masked_mapdata, vmin=vmin, vmax=vmax, cmap=cmapname, edgecolor="#aaaaaa", linewidth=0.5)
    
    plot_text(targetData, blockedMap, whitelabelmin)
    
    for path in lines:
        ys, xs = zip(*path)
        plt.plot(xs, ys, c="black", lw=1.5)
    
    plt.gca().invert_yaxis()
    plt.axis("off")
    plt.tight_layout()
    cb = plt.colorbar(shrink=0.1, aspect=10)
    cb.set_label(targetData)
    plt.show()
```
