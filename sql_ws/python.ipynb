## 4. Fetch All

```python
connect = mysql.connector.connect(
    host='database-1.cd08osauaz8j.us-east-2.rds.amazonaws.com',
    port=3306,
    user='admin',
    password='ghkcjs123',
    database = 'zerobase'
)
cur = connect.cursor(buffered=True)
cur.execute('SELECT * FROM sql_file')

result = cur.fetchall()
for result_iterator in result:
    print(result_iterator)

connect.close()
```

## 5. csv에 있는 데이터를 python으로 INSERT 예제

```python
police_df = pd.read_csv('police_station.csv')
police_df.head()
```

```python
# police_df에서 zerobase 데이터베이스의 police_station에 데이터 입력
connect = mysql.connector.connect(
    host='database-1.cd08osauaz8j.us-east-2.rds.amazonaws.com',
    port=3306,
    user='admin',
    password='ghkcjs123',
    database = 'zerobase'
)
cur = connect.cursor(buffered=True)
sql = 'INSERT INTO police_station VALUES (%s, %s)'

for i, row in police_df.iterrows():
    cur.execute(sql, tuple(row))
    print(tuple(row))
    # commit()은 데이터베이스에 적용하기 위한 명령
    connect.commit()
```

```python
# 가져온 데이터 확인
cur.execute('SELECT * FROM police_station')
result = cur.fetchall()
for row in result:
    print(row)

connect.close()
```

```python
# 데이터프레임으로 가져온 데이터 넣기
station_df = pd.DataFrame(result)
station_df
```
